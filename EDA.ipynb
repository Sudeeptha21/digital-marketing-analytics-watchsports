```python
import pandas as pd
import numpy as np
```

```python
channel_perf = pd.read_csv(r"C:\Users\91744\Desktop\projects\CrimsonProject\channel_performance_202508091723.csv")
viewer_behavior = pd.read_csv(r"C:\Users\91744\Desktop\projects\CrimsonProject\viewer_behavior_202508091729.csv")
vid_perf = pd.read_csv(r"C:\Users\91744\Desktop\projects\CrimsonProject\top_video_performance_materialized_202508091737.csv")
platform_act = pd.read_csv(r"C:\Users\91744\Desktop\projects\CrimsonProject\platform_daily_activity_202508091723.csv")
```

```python
channel_perf.head()
channel_perf.info()
channel_perf.describe()

```

```python
print(viewer_behavior.head())
#impute missing values in total_watch_time to 0
viewer_behavior['total_watch_time'] = viewer_behavior['total_watch_time'].fillna(0)
# impute missing values in country to unknown
viewer_behavior['country'] = viewer_behavior['country'].fillna('Unknown')
print(viewer_behavior.info())
print(viewer_behavior.describe())
```

```python
print(vid_perf.head())
print(vid_perf.describe())
# fill the missing values in total_view, total_engagements, avg_playback_position, avg_watch_percent to 0
vid_perf['total_views'] = vid_perf['total_views'].fillna(0)
vid_perf['total_engagements'] = vid_perf['total_engagements'].fillna(0)
vid_perf['avg_playback_position'] = vid_perf['avg_playback_position'].fillna(0)
vid_perf['avg_watch_percent'] = vid_perf['avg_watch_percent'].fillna(0)
print(vid_perf.info())
```

```python
print(platform_act.head())
print(platform_act.info())
print(platform_act.describe())
```

```python
#write to this folder C:\Users\91744\Desktop\projects\CrimsonProject\edited_Csvs
channel_perf.to_csv("channel_performance_clean.csv", index=False)
viewer_behavior.to_csv("viewer_behavior_clean.csv", index=False) 
vid_perf.to_csv("video_performance_clean.csv", index=False)   
platform_act.to_csv("platform_activity_clean.csv", index=False)

print("Cleaned files saved successfully.")
```
